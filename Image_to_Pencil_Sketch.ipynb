**AN IMAGE TO PENCIL SKETCH WITH PYTHON.**

STEPS.

1. Find an image to convert to a pencil sketch.

2. Read the image in RBG format and  convert it to a grayscale image. This will turn an image into a classic black and white photo.

3. Invert the grayscale image  (negative image), this will be our inverted grayscale image. 

4.  Create the pencil sketch by mixing the grayscale image with the inverted blurry image.

**IMPORTING LIBRARIES.**

The library to be used is OpenCV.

```python
#Importing Libraries.
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from PIL import Image
```

```python
#Reading the image
image = io.imread("/content/drive/MyDrive/LGMVIP/Image_to_Pencil_Sketch/Messi.jpeg")
cv2_imshow(image)
```

```python
#Operation performed on the image.
grey_filter = cv.cvtColor(image , cv.COLOR_BGR2GRAY)
cv2_imshow(grey_filter)
```

```python
#Inverting the gray scale image.
invert = cv.bitwise_not(grey_filter)
cv2_imshow(invert)
```

```python
#BLuring the image.
blur = cv.GaussianBlur(invert, (21,21) ,0)
cv2_imshow(blur)
```

```python
#Inverting the blur image.
invertedblur = cv.bitwise_not(blur)
cv2_imshow(invertedblur)
```

```python
sketch_image = cv.divide(grey_filter , invertedblur , scale =255.0)
cv2_imshow(sketch_image)
cv.imwrite('Sketch.jpg' , sketch_image)
```
